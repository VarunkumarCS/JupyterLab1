# Linearly Spaced Points
There are many tasks that you need to generate a set of points in a range that are linearly spaced. For this task
we will use numpy modules arnage and linspace.

Note that the tutorial uses `np.arange`. While this is fine in some cases, we generally prefer to use `np.linspace`. `np.linspace(a, b, N)` divides the interval `[a, b]` into N points, while `np.arange(a, b, s)` will step from `a` to `b` with a fixed step size `s`.

One thing to keep in mind is that `np.linspace` always includes both end points while `np.arange` will *not* include the second end point `b`. For this reason, when we are plotting ranges of values we tend to prefer `np.linspace`.

Write two functions that return the same result.

```python
# create an array of linearly spaced points from -5(inclusive) to 6 (not inclusive) spaced by 1
d = np.arange(-5,6,1)
print(d)

# do the same with linspace
e=np.linspace(-5,6,12)
print(e)
```

## Question 2.3 3D Plotting
The code below plot function f(x,y) in 3D. The function is 
$$f(x,y) = x^2 + y^2$$
### Activity 1
Observe the values of (x,y) that minimizes the function f(x,y)

Function f(x,y) is returning the sum of squares of x and y. As, x and y are squaring in the function, all the negative values in the function will be also returning the positive answer. Therefore, the minimum value will be (0,0) for x and y.

### Activity 2
Change the function to $$f(x,y) = xy $$ and re-plot. Estimate the values (x,y) that minimizes f(x,y)?

Function (x,y) is returning the product of x and y. 
The function will return the minimum value when the x and y have the opposite sign. 
This means when y = negative, x will be positive and vice versa ( x = negative, y will be positive). 


```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return (x*y)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

## Task 3 - Reading Documentation
We should also familiarize ourselves with looking up documentation and learning how to read it. Below is a section of code that plots a basic wireframe. Replace each `# Your answer here` with a description of what the line above does, what the arguments being passed in are, and how the arguments are used in the function. For example your answer may look like,

`np.arange(2, 5, 0.2)`

`# This returns an array of numbers from 2 to 5 with an interval size of 0.2`

Hint: The `Shift` + `Tab` (to see what a function does) tip from earlier in the notebook may help here.

```python
from mpl_toolkits.mplot3d import axes3d

u = np.linspace(1.5*np.pi, -1.5*np.pi, 100)
# np.linespace() is a function that returns number spaces evenly w.r.t interval. It's similar to numpy.arange() function but instead of step it uses a sample number.
# The above line returns evenly spaced samples which are 100 , 1.5*np.pi is the starting point and -1.5*np.pi is the stopping point.

[x,y] = np.meshgrid(u, u)
# This function is used to create a rectangular grid out of an array of x values and an array of y values. 
# The above line means that ordered pairs of (x,y) coordinates will begin from range starting points of u and will end with ending points of u.

squared = np.sqrt(x.flatten()**2 + y.flatten()**2)
z = np.cos(squared)
# The above line computes (x^2+y^2)^(1/2)

z = z.reshape(x.shape)
# The above line is changing the dimension of z, without changing its data. 

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# After creating the figure, the line helps to create 3D axis object by calling add_subplot to the figure and specifying the value '3d. to the projection parameter. Axis object is used as 'ax' to add any plot to the figure.
# The number '111' is the way of locating subplot in the overall space of the figure object.

ax.plot_wireframe(x, y, z, rstride=10, cstride=10)
# The above line creates the 3d surface plots.
# The keywords "rstride" and "cstride" determine the row step size and the column step size. these keywords arguments control how close together the "wires" in the wire frame plot are drawn.

ax.view_init(elev=50., azim=30)
# The above line sets the view point, so the viewer can see. This can be used to rotate the axes programmatically.
# 'elev'stores the elevation angle in the z(elevated angle) and 'azim' stores the azimuth angle in the x,y(horizontol angle)plane.
plt.savefig("figure1.png")
```
